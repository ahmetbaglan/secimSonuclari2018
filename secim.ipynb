```python
import pandas as pd 


%matplotlib inline
```

```python
sonuclar = pd.read_csv('secimData2018Clean.csv')
```

```python
partiler = ['AKP',  'CHP', 'HDP','İYİ', 'MHP', 'SP','HÜDA-PAR','VP','BGMSZ']

cb_adaylari = ['Recep Tayyip Erdoğan', 'Muharrem Ince',
       'Meral Akşener', 'Selahattin Demirtaş', 'Temel Karamollaoğlu','Doğu Perinçek']

gerekli_bilgi = ['Il', 'Ilce','Toplam Seçmen']
```

# CB adaylarının en çok oy aldığı ilçeler

## Recep Tayyip Erdoğan

```python
sonuclar[gerekli_bilgi + cb_adaylari].sort_values(by = 'Recep Tayyip Erdoğan', ascending = False)
```

Gerçekten ekstrem ilçeler var. %90 üzeri oy veren 7 ilçe var. 

## Muharrem Ince

```python
sonuclar[gerekli_bilgi + cb_adaylari].sort_values(by = 'Muharrem Ince', ascending = False)
```

Aslında görüldüğü üzere İstanbul'da Beşiktaş ve Kadıköy en ekstrem noktalardan. Bu bölgelerin toplam oy potansiyeli %1.

## Selahattin Demirtaş 

Seçimdeki en ilginç aktörlerdendi Demirtaş. Benzer bir istatistiği onun için yapacak olursak aşağıdaki sonuçla karşılaşıyoruz.

```python
sonuclar[gerekli_bilgi + cb_adaylari].sort_values(by = 'Selahattin Demirtaş', ascending = False)
```

Aslında benim merak ettiğim bir konu da acaba hangi bölgeler de parti olarak HDP tercih edilirken CB başkanı adayı olarak Selahattin Demirtaş seçilmedi (Bir oy HDP'ye bir oy İnce'ye propagandası) Bunu tam olarak saptamak zor ama fark alıp sıralarsak fikir verici olabilir.

```python
hdpSeloSonuclar = sonuclar[gerekli_bilgi + ['Selahattin Demirtaş', 'HDP']]
hdpSeloSonuclar['fark'] = hdpSeloSonuclar['HDP'] - hdpSeloSonuclar['Selahattin Demirtaş']
hdpSeloSonuclar.sort_values(by = 'fark', ascending = False)
```

Burada  CHP'nin yüksek oy aldığı yerlerde daha büyük bir fark olduğu görülebilir. 

Yapılabilcek bir diğer şey ise HDP ve Selahattin Demirataşa gelen oyları Scatter Plot şeklinde göstermek 

```python
sonuclar[['HDP','Selahattin Demirtaş']].plot.scatter(x = 'HDP', y = 'Selahattin Demirtaş')
```

 Görüldüğü üzere çok büyük bir korelasyon var.
